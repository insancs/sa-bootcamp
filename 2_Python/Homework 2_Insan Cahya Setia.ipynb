Siapakah pemain yang memiliki warganegara 'France' dan berada pada Club 'Manchester United’?

```python
# condition for players with French nationality
cond1 = (df_fifa['Nationality'] == 'France')
# condition for players with club is Manchester United
cond2 = (df_fifa['Club'] == 'Manchester United')
df_fifa[cond1 & cond2]
```

Tampilkan 'Name', 'Age', 'Overall', 'Wage', dan 'Release Clause' pada pemain dengan kewarganegaraan 'England' dan Club 'Manchester City’

```python
# condition for players with England nationality
cond1 = (df_fifa['Nationality'] == 'England')
# condition for players with club is Manchester City
cond2 = (df_fifa['Club'] == 'Manchester City')

df_fifa[cond1 & cond2].loc[:, ['Name', 'Age', 'Overall', 'Wage', 'Release Clause']].reset_index(drop=True)
```

Lakukan pengelompokkan Club (groupby Club) dan tampilkan rata-rata overall pemain.

```python
# Average Overall for each Club 
df_fifa.groupby('Club')['Overall'].mean().reset_index()
```

Lakukan pengelompokkan berdasarkan Nationality dan tampilkan jumlah pemain dan median umur.

```python
nationality_group = df_fifa.groupby('Nationality').agg({'Name':'count', 'Age':'median'}).reset_index()
nationality_group = nationality_group.rename(columns={'Name':'Jumlah Pemain', 'Age':'Median Umur'})
nationality_group
```

Membuat kolom baru bernama 'segment_player', yang memuat status segmentasi pemain sbb :
1. 'High' jika overall pemain > 85
2. 'Mid' jika overall pemain antara 65 dan 85
3. 'Low' jika overall pemain dibawah 65

```python
df_fifa['segment_player'] = np.select([df_fifa['Overall'] > 85, # more than 85 for high 
                                       df_fifa['Overall'] >= 65, # between 65 and 85 for mid                                 
                                       df_fifa['Overall'] < 65], # lower than 65 for low
                                      ['High', 'Mid', 'Low'])

# Show 20 samples data
df_fifa[['Overall', 'segment_player']].sample(20)
```

Urutkan data berdasarkan Overall secara Descending.

```python
df_fifa.sort_values('Overall', ascending=False)
```

# Homework 2 - Merging Data

## Load US Videos Dataset

```python
# Load US Videos 
df_videos = pd.read_csv(path + 'USvideos.csv')
df_videos.head()
```

```python
df_videos.info()
```

Buat dataframe dengan nama 'title' yang berisi video_id dan title.

```python
title = df_videos[['video_id', 'title']]
title.head()
```

Buat dataframe dengan nama 'description' yang berisi video_id dan description

```python
description = df_videos[['video_id', 'description']]
description.head()
```

Export keduanya ke dalam CSV baru dan import Kembali menggunakan glob

```python
# Export title and desription dataframe to csv file
title.to_csv('title.csv', index=False)
description.to_csv('description.csv', index=False)

# Import glob to read multiple csv files
from glob2 import glob
file_name = glob('*.csv')

# Create list
dataframes = []

for file in file_name:
  df = pd.read_csv(file)
  dataframes.append(df)
```

```python
dataframes[1].head(3)
```

```python
dataframes[0].head(3)
```

Lakukan operasi merge (boleh dengan .merge atau .join) kepada dataframe title dan description. kemudian munculkan title video yang tidak ada descriptionnya.

```python
# merge title and description dataframes
title_desc = pd.merge(dataframes[1], dataframes[0], on='video_id', how='inner')
title_desc
```

```python
# Filter for descriptions that have null values
desc_null = title_desc[title_desc['description'].isnull()].reset_index(drop=True)
desc_null
```

Ada berapa nilai unik title yang tidak ada descriptionnya?

```python
# Number of unique values for the title
print('Nilai unik title yang tidak mempunyai description ada : {}'.format(desc_null['title'].nunique()))
```
