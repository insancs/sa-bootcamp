## Matplotlib

#### Basic Plot

```python
x = np.linspace(0, 10, 10)
y = x**2
z = x + y
```

```python
plt.plot(x, y)
plt.title('My First Chart')
plt.xlabel('Sumbu X')
plt.ylabel('Sumbu Y')
plt.show()
```

#### Subplot

```python
plt.subplot(2, 3, 1)
plt.plot(x, y)

plt.subplot(2, 3, 2)
plt.plot(x, x)

plt.subplot(2, 3, 3)
plt.plot(y, x)

plt.subplot(2, 3, 4)
plt.subplot(2, 3, 5)
plt.subplot(2, 3, 6)
plt.tight_layout() # merapikan layout
```

#### Plot Type

```python
plt.subplot(2, 3, 1)
plt.scatter(x, y)

plt.subplot(2, 3, 2)
plt.plot(x, x)

plt.subplot(2, 3, 3)
plt.hist(x, y)
plt.tight_layout() # merapikan layout
```

#### Legend

```python
plt.subplot(1, 2, 1)
plt.plot(x, y, label='Garis 1')
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(y, x, label='Garis 2')
plt.legend()

plt.show()
```

```python
plt.subplot(1, 2, 1)
plt.plot(x, y, label='Garis 1')
plt.plot(y, x, label='Garis 2')
plt.legend()
```

#### Object Oriented Plot

```python
fig = plt.figure()
axes = fig.add_axes([0, 0, 0.8, 0.5]) # [x0, y0, x1, y1]
axes.plot(x, y)
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0, 0, 1, 1]) # [x0, y0, x1, y1]
axes2 = fig.add_axes([0.5, 0.5, 1, 1]) # [x0, y0, x1, y1]
axes1.plot(x, y)
axes2.scatter(x, y)
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0, 0, 1, 1]) # [x0, y0, x1, y1]
axes2 = fig.add_axes([0, 0.5, 0.5, 0.5]) # [x0, y0, x1, y1]
axes1.plot(x, y)
axes2.scatter(x, y)
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0, 0, 1, 1]) # [x0, y0, x1, y1]
axes2 = fig.add_axes([1.2, 0, 1, 1]) # [x0, y0, x1, y1]
axes1.plot(x, y)
axes2.scatter(x, y)
```

Object Oriented Using Subplots

```python
fig, ax = plt.subplots(1, 2)
ax[0].plot(x, y)
ax[1].scatter(x, y)
```

```python
fig, ax = plt.subplots(2, 2)
ax[0, 0].plot(x, y)
ax[0, 1].scatter(x, y)
ax[1, 0].hist(x, y)
ax[1, 1].fill(x, y)
plt.tight_layout()
```

#### Figure Size, Title, Axes

Figure Size in Object Oriented

```python
fig = plt.figure(figsize=(15, 7))
axes = fig.add_axes([0, 0, 0.5, 0.5]) # [x0, y0, x1, y1]
axes.plot(x, y)
axes.set_title('Plot X dan Y')
axes.set_xlabel('Sumbu X')
axes.set_ylabel('Sumbu Y')
plt.show()
```

```python
fig = plt.figure(figsize=(3, 3))

axes_1 = fig.add_axes([0,0,1,1])
axes_2 = fig.add_axes([1.2, 0.5, 0.5, 0.5])

axes_1.plot(x, y)
axes_2.scatter(x, y)

axes_1.set_title('Line Chart')
axes_1.set_xlabel('sumbu-X')
axes_1.set_ylabel('sumbu-Y')

axes_2.set_title('Scatter Chart')
axes_2.set_xlabel('X')
axes_2.set_ylabel('y')
```

```python
fig, axes = plt.subplots(2, 2, figsize=(10, 8))
axes[0, 0].plot(x, y)
axes[0, 0].set_title('Line Plot X dan y')

axes[0, 1].scatter(x, y)
axes[0, 1].set_xlabel('Sumbu X')

axes[1, 0].hist(x, y)
axes[1, 0].set_ylabel('Sumbu y')

axes[1, 1].fill(x, y)
axes[1, 1].set_xlabel('Sumbu X')
axes[1, 1].set_ylabel('Sumbu y')
plt.tight_layout()
```

## Seaborn

### Distribution Plot

```python
diamond = sns.load_dataset('diamonds')
diamond.head()
```

```python
sns.displot(data=diamond, x='price', kde=True)
plt.show()
```

```python
sns.distplot(diamond['price'])
plt.show()
```

#### Rugplot

```python
sns.kdeplot(data=diamond, x='price')
sns.rugplot(data=diamond, x='price')
```

#### Violin Plot

```python
sns.violinplot(data=diamond, x='price')
```

### Categorical Plot

#### Composition

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
sns.countplot(data=tips, x='sex')
```

```python
sns.countplot(data=tips, x='day')
```

```python
jumat = tips[tips['day'] == 'Fri']
sns.countplot(data=jumat, x='time', hue='sex')
```

#### Comparison

Residal

```python
sns.residplot(data=tips, x='total_bill', y='tip')
```

Joint Plot

```python
sns.jointplot(data=tips, x='total_bill', y='tip', kind='hex')
```

## Superstore

```python
pd.set_option('display.max_columns', 30)
```

```python
df_store = pd.read_excel('data/Superstore.xls')
df_store.head()
```

```python
plt.figure(figsize=(12, 8))
df_store.groupby('Sub-Category')['Sales'].sum().sort_values(ascending=False).plot(kind='bar')
```

```python
df_store_2016 = df_store[(df_store['Order Date'] >= '2016-01-01') & (df_store['Order Date'] <= '2016-12-31')]
df_store_2016['month'] = df_store_2016['Order Date'].dt.month
df_store_2016
```

```python
plt.figure(figsize=(12, 8))
df_store_2016.groupby(['month', 'Category'])['Sales'].sum().unstack().plot(kind='line')
plt.title('Sales per Month in 2016', fontsize=20, fontweight='bold', pad=15)
plt.xlabel('Month', fontsize=15)
plt.ylabel('Sales', fontsize=15)
plt.tight_layout()
```
