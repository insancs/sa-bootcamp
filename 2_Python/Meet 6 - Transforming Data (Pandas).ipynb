## Merging

```python
product_year = pd.DataFrame({'year':[2013, 2014, 2015, 2016, 2017, 2018], 
                             'total_product': [1234, 2341, 4322, 2344, 1231, 3435]})

profit_year = pd.DataFrame({'year':[2015, 2016, 2017, 2018, 2019, 2020], 
                            'total_profit': [100, 200, 300, 400, 500, 699]})
product_year
```

```python
profit_year
```

```python
pd.merge(product_year, profit_year)
```

```python
pd.merge(product_year, profit_year, how='left')
```

```python
pd.merge(product_year, profit_year, how='outer')
```

```python
product_year = pd.DataFrame({'year':[2013, 2014, 2015, 2016, 2017, 2018], 
                     'total_product': [1234, 2341, 4322, 2344, 1231, 3435],
                             'sales': [423, 232, 332, 131, 332, 121]})

profit_year = pd.DataFrame({'year':[2015, 2016, 2017, 2018, 2019, 2020], 
                     'total_profit': [100, 200, 300, 400, 500, 699],
                           'sales': [43422, 54533, 23232, 3434, 45554, 65656]})

product_year
```

```python
profit_year
```

```python
pd.merge(product_year, profit_year, left_on='year', right_on='year', suffixes=('_product', '_profit'))
```

Merge_ordered merupakan syntax merge yang digunakan untuk menggabungkan dua dataframe dengan pengisian data NaN secara otomatis.

```python
product_year = pd.DataFrame({'year':[2013, 2014, 2015, 2016, 2017, 2018], 
                             'total_product': [1234, 2341, 4322, 2344, 1231, 3435]})

profit_year = pd.DataFrame({'year':[2015, 2016, 2017, 2018, 2019, 2020], 
                            'total_profit': [100, 200, 300, 400, 500, 699]})
```

```python
pd.merge_ordered(product_year, profit_year, fill_method='ffill')
```

## Exercise

```python
registrasi = pd.DataFrame({'reg_id':[1, 2, 3, 4],
                          'nama':['Andi', 'Budi', 'Ujang', 'Wawan'], 
                          'phone':['08374682726', '089736746528', '08364729273', '08736453290']})
login = pd.DataFrame({'log_id':[1, 2, 3, 4],
                          'nama':['Boy', 'Budi', 'David', 'Wawan'], 
                          'phone':['083746098026', '089736746528', '08876586273', '08736453290']})

registrasi
```

```python
login
```

1. Tampilkan data seluruh peserta yang registrasi dan hadir dalam satu dataframe. berikut adalah expected resultnya

```python
all_data = pd.merge(registrasi, login, how='outer')
all_data
```

2. kamu menyadari ada beberapa yang hadir dan belum registrasi. Tampilkan nama dan nomor hape peserta yang hadir dan belum registrasi untuk diminta registrasi terlebih dahulu

```python
log_noreg= pd.merge(registrasi, login, how='right')
log_noreg = log_noreg[log_noreg['reg_id'].isnull()].iloc[:, 1:]
log_noreg
```

3. Kamu juga menyadari ada beberaa yang sudah registrasi namun belum hadir. kamu ingin menghubungi mereka dan melakukan konfirmasi kehadiran. munculkan data peserta yang sudah registrasi tapi belum hadir.

```python
reg_nolog = pd.merge(login, registrasi, how='right')
reg_nolog = reg_nolog[reg_nolog['log_id'].isnull()].iloc[:, 1:]
reg_nolog
```

4. Kamu ingin mengetahui persentase :
- jumlah peserta yang hadir dan sudah registrasi (dari total keseluruhan),
- Jumlah peserta yang registrasi namun tidak hadir (dari total peserta yang regis), dan
- Jumlah peserta yang hadir namun tidak registrasi (dari total peserta yang hadir)

```python
reg_log =  len(all_data[(all_data['reg_id'].isnull()==False) & (all_data['log_id'].isnull()==False)])
pct_reg_log = reg_log/len(all_data)

# persentase peserta regist namun tidak login
pct_reg_nolog = len(reg_nolog)/len(registrasi)

# persentase peserta login namun tidak regist
pct_log_noreg = len(log_noreg)/len(login)
```

```python
print('Persentase peserta yang sudah regist dan hadir : {:.2%}:'.format(pct_reg_log))
print('Persentase peserta yang regist namun tidak hadir : {:.2%}'.format(pct_reg_nolog))
print('Persentase peserta yang login namun tidak regist : {:.2%}'.format(pct_log_noreg))
```
